```python
import pandas as pd
import os
import shutil
```

```python
df = pd.read_csv('data\\train-metadata.csv')
df
```

```python
ids = df[df.target == 1 ].isic_id
ids
```

# working on oversampling

```python
src_path= "data\\train-image\\image\\"
dest_path = "data\\train-image\\image\\oversampledCancer\\"
```

```python
for id in ids:
    print("Source: "+ src_path + id)
    print("Destination: "+ dest_path + id)
    shutil.move(src_path + id+ ".jpg", dest_path + id + ".jpg")
```

# Wroking on undersampling

```python
ids = df[df.target == 0 ].isic_id
ids = ids.sample(10000)
ids
```

```python
src_path= "data\\train-image\\image\\normal\\"
dest_path = "data\\train-image\\image\\undersampledNormal\\"

os.makedirs(dest_path, exist_ok=True)
```

```python
for id in ids:
    print("Source: "+ src_path + id)
    print("Destination: "+ dest_path + id)
    shutil.move(src_path + id+ ".jpg", dest_path + id + ".jpg")
```

```python
images_final_location = "data\\train-image\\image\\images\\"

os.makedirs(images_final_location, exist_ok=True)
```

```python
dest_path = "data\\train-image\\image\\undersampledNormal\\"
normal_images = os.listdir(dest_path)

for image in normal_images:
    shutil.move(dest_path + image, images_final_location + image)
```

```python
dest_path = "data\\train-image\\image\\oversampledCancer\\"
cancer_images = os.listdir(dest_path)

for image in cancer_images:
    shutil.move(dest_path + image, images_final_location + image)
```
